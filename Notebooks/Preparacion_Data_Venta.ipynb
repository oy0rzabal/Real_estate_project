#Limpieza de los datos de Inmubele 24.

## Importamos las libreriras necesarias para empezar con la limpieza de datos.

```python
import pandas as pd
import numpy as np
```

## Mandamos atraer los datos conseguidos de ambas paginas

```python
inmu_venta = 'Inmuebles_venta.xlsx'
Propi_venta = 'Propiedades_venta.xlsx'

in_venta = pd.read_excel(inmu_venta)
Propi_venta = pd.read_excel(inmu_venta)

Propi_venta = pd.DataFrame(Propi_venta)
in_venta = pd.DataFrame(in_venta)
```

#Exploracion de Inmueble24

```python
in_venta.head(1)
```

##### Empezamos a eliminar las columnas inncesarias.

```python
in_venta = in_venta.drop(['gotoposting','Info','Image', 'Image1', 'Image2', 'Image3', 'Image4', 'Image5', 'Info6', 'Image7', 'postingpills', 'gotoposting_URL', 'publisherlogo_URL', 'Image8'], axis=1)
in_venta.head(1)
```

#### Empezamos a cambiar los nombres de las columnas.


```python
in_venta = in_venta.rename(columns={'Title':'Ubicacion','Price':'Precio','Location':'Colonia','Keywords':'M2','Description':'Descripcion'})
in_venta.head(1)
```

#### Analisis profundo en Precio

```python
in_venta.dtypes
```

##### Comprobamos que tiene MN Y USD, lo que haremos es separalos en una nueva columna y borrar la columna con los datos inncesarios

```python
in_venta['Precio'].head()
```

```python
in_venta1 = in_venta['Precio'].str.split(expand=True)
in_venta1.head(1)
```

#### Comprobamos que tipo de datos es "Precio" para poder continuar con nuestro analisis, si vemos que es tipo object lo pasaremos a str hasta llegar a tipo entero o a flotante

```python
in_venta1.dtypes
```

#### Convertimos los object a entero

```python
in_venta1[1] = in_venta1[1].astype(int)
```

```python
in_venta1.dtypes
```

#### Ahora vamos a quitar cualquier dato innecesario que tengamos en nuestra tabla para darle el toque final.

```python
in_venta['M2'].head(2)
```

```python
M2 = in_venta['M2'].str.split("\n", n=2, expand=True)
M2 = in_venta['M2'].str.split(expand=True)
M2.head(2)
```

#### Comenzamos con la union de los datos con un nuevo DataFrame llamado venta

```python
venta = pd.DataFrame()
```

### Empezamos a reestructurar todo el data frame hasta crear uno nuevo

```python
#Unimos las columnas
venta['Descripcion'] = in_venta['Ubicacion']
venta['Ubicacion'] = in_venta['Colonia']
venta = venta.assign(Colonia='CanutoNogueda')
venta = venta.assign(Zona='Diamante')
venta['USD/MX'] = in_venta1[0]
venta['Numero_Total'] = in_venta1[1]
venta['M2'] = M2[1]
venta['Numero_m2'] = M2[0]
venta['Cantidad'] = M2[2]
venta['Banos/Estacionamiento'] = M2[3]
```
